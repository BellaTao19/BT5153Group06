```python
df = df.str.lower()
df = df.str.replace('[^\w\s]', ' ')

# how else should the text be processed?
df = df.str.replace("'s'", ' ')
from string import digits
remove_digits  = str.maketrans('','',digits)
df = df.str.translate(remove_digits)
df.head()
```

```python
df_copy = df.copy()
x_temp = df_copy.to_list()
```

```python
x_temp[1]
```

```python
x=[]
for i in x_temp:
    if len(str(i))>0:
        x.append(str(i))
#y = df_copy.Outcome.to_list()

#train_x,test_x,train_y,test_y = train_test_split(x, y, test_size=0.2, random_state=0, stratify=y)
```

```python
stop_words=["a","an","and","are","as","at","be","by","for","from","has","he","in","is","it","its","of","on","that","the","to","was","were","will","with"]

cv = CountVectorizer(stop_words='english', ngram_range=(1,1), min_df=2, max_df=0.5, max_features=None, token_pattern="[a-z]+")

train_x_counts = cv.fit_transform(x)

tfidf = TfidfTransformer(sublinear_tf=True)
train_x_tfidf = tfidf.fit_transform(train_x_counts)
```

```python
a= train_x_counts
a_name = cv.get_feature_names()
len(a_name)
```

```python
sum_words = np.array(a.sum(axis=0))[0]
```

```python
sum_words
len(sum_words)
```

```python
word_frq = {}
for i in range(len(a_name)):
    word_frq[a_name[i]] = sum_words[i]

sort_dic = {k: v for k, v in sorted(word_frq.items(), key=lambda item: item[1], reverse = True)}
```
